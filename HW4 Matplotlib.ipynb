# Задание 2

```python
t  = np.linspace(0, 10, 51)
t
```

```python
x = np.cos(t)
x
```

```python
plt.plot(x, t)
plt.title("График f(t)")
plt.xlabel("Значения t")
plt.ylabel("Значения x")

plt.show()
```

```python
plt.plot(x, t)
plt.title("График f(t)")
plt.xlabel("Значения t")
plt.ylabel("Значения x")
plt.axis([0.5, 9.5, -2.5, 2.5])

plt.show()
```

# Задание 3

```python
x  = np.linspace(-3, 3, 51)
x
```

```python
y1 = x**2
y1
```

```python
y2 = y2 = 2 * x + 0.5
y2
```

```python
y3 = -3 * x - 1.5
y3
```

```python
y4 = np.sin(x)
y4
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)
plt.show()
```
